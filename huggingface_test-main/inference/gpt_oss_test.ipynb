## pipeline 활용 추론 테스트

### gpt-oss-20B

```python
from transformers import pipeline

generator = pipeline(
    "text-generation",
    model="openai/gpt-oss-20b",
    torch_dtype="auto",
    device_map="auto"  # Accelerate 라이브러리 이용 멀티 GPU 사용(모델 레이어를 여러 GPU에 분산 배치)
)

messages = [
    {"role": "user", "content": "안녕 너는 어떤 모델이고 어떤걸 할 수 있어?"},
]

result = generator(
    messages,
    max_new_tokens=200,
    temperature=1.0,
)

print(result[0]["generated_text"])
```
